```python
import sqlite3, pandas as pd, numpy as np
from sentence_transformers import SentenceTransformer

conn = sqlite3.connect(r"C:\Users\Gebruiker\Desktop\LOCAL_SITE\ungdc-explorer\backend\ungdc.db")
df_topics = pd.read_csv("table.csv", encoding="utf-8")

model = SentenceTransformer("all-MiniLM-L6-v2")
# use GPU if available
device = "cuda" if model.device.type == "cuda" else "cpu"
print("Using:", device)

topic_texts = df_topics["Short description (for embeddings / app)"].tolist()
topic_ids = df_topics["topic_final"].tolist()
topic_labels = df_topics["New label (suggested)"].tolist()
topic_embs = model.encode(topic_texts, normalize_embeddings=True)

def chunk_text(text, words_per_chunk=150, overlap=30):
    words = text.split()
    chunks = []
    start = 0
    while start < len(words):
        end = start + words_per_chunk
        chunks.append(" ".join(words[start:end]))
        start += words_per_chunk - overlap
        if end >= len(words):
            break
    return chunks

speeches = pd.read_sql("SELECT rowid as speech_id, text FROM speeches", conn)

# build a flat list of all chunks with back-references
all_chunks = []
chunk_map = []  # (speech_id, chunk_idx)
for _, row in speeches.iterrows():
    chunks = chunk_text(row["text"])
    for i, c in enumerate(chunks):
        all_chunks.append(c)
        chunk_map.append((row["speech_id"], i))

print("Total chunks:", len(all_chunks))

# batch encode everything at once (this is the slow step, but efficient)
chunk_embs = model.encode(all_chunks, normalize_embeddings=True, batch_size=64, show_progress_bar=True)

# similarity: (n_chunks, n_topics)
sims = chunk_embs @ topic_embs.T

TOP_K = 3
THRESHOLD = 0.2

rows_to_insert = []
for j, (speech_id, chunk_idx) in enumerate(chunk_map):
    sim_row = sims[j]
    top_idx = np.argsort(sim_row)[::-1][:TOP_K]
    for idx in top_idx:
        if sim_row[idx] >= THRESHOLD:
            rows_to_insert.append((
                int(speech_id), chunk_idx,
                int(topic_ids[idx]), topic_labels[idx],
                float(sim_row[idx])
            ))

cur = conn.cursor()
cur.execute("DROP TABLE IF EXISTS chunk_topics_new")
cur.execute("""
CREATE TABLE chunk_topics_new (
    speech_id INTEGER,
    chunk_idx INTEGER,
    topic_final INTEGER,
    topic_label TEXT,
    score REAL
)
""")
cur.executemany(
    "INSERT INTO chunk_topics_new (speech_id, chunk_idx, topic_final, topic_label, score) VALUES (?,?,?,?,?)",
    rows_to_insert
)
conn.commit()
```

```python
import sqlite3
import pandas as pd
import numpy as np
from sentence_transformers import SentenceTransformer
from pathlib import Path
```

```python
conn = sqlite3.connect(r"C:\Users\Gebruiker\Desktop\LOCAL_SITE\ungdc-explorer\backend\ungdc.db")
cur = conn.cursor()
```

```python
speech_topics = pd.read_sql("SELECT * FROM speech_topics_new", conn)

main_topic = speech_topics.sort_values(["speech_id","n_chunks","avg_score"], ascending=[True,False,False]) \
                           .groupby("speech_id").first().reset_index()

main_topic.to_sql("speech_main_topic_new", conn, if_exists="replace", index=False)
conn.commit()
print("speech_main_topic_new created, rows:", len(main_topic))
```

```python
# load main topic per speech
main_topic = pd.read_sql("SELECT * FROM speech_main_topic_new", conn)

# join with speeches to get iso/year for matching
speech_ids = pd.read_sql("SELECT rowid as speech_id, iso, year FROM speeches", conn)
main_topic = main_topic.merge(speech_ids, on="speech_id")

cur = conn.cursor()
for _, row in main_topic.iterrows():
    cur.execute("""
        UPDATE speeches 
        SET topic_final = ?, topic_final_name = ?
        WHERE iso = ? AND year = ?
    """, (int(row["topic_final"]), row["topic_label"], row["iso"], int(row["year"])))

conn.commit()
print("speeches updated")
```
