```python
from bertopic import BERTopic
from pathlib import Path

MODEL_PATH = Path(r"C:\Users\Diaz Wiersma\Desktop\AP\SysteemProjectDelen\dataverse_files\bertopic_model")

print("Model laden...")
model = BERTopic.load(str(MODEL_PATH))
print("Geladen.")

# Herclassificeer outliers
print("Outliers herclassificeren...")
new_topics = model.reduce_outliers(
    df["text_clean"].tolist(),
    df["topic"].tolist(),
    strategy="c-tf-idf",  # sneller dan "embeddings", geen GPU nodig
    threshold=0.1
)

df["topic_reduced"] = new_topics
df["topic_reduced_name"] = df["topic_reduced"].map(topic_name_map)

print("Outliers voor:", (df["topic"] == -1).sum())
print("Outliers na:", (df["topic_reduced"] == -1).sum())
print("\nVerdeling nieuwe topics:")
print(df["topic_reduced_name"].value_counts().head(10))
```

```python
print("Top topics (reduced):")
print(df["topic_reduced"].value_counts().head(10))

print("\nAantal outliers (-1):", (df["topic_reduced"] == -1).sum())
```

```python
# Werk de topic kolommen bij met de gereduceerde versie
df["topic_final"] = df["topic_reduced"]
df["topic_final_name"] = df["topic_reduced"].map(topic_name_map)

# Gooi kolommen weg die de backend niet nodig heeft
df = df[[
    "iso", 
    "country", 
    "year", 
    "session",
    "speaker",
    "post",
    "text",
    "topic_final",
    "topic_final_name"
]].copy()

# Fix de 80 nulls in country
df["country"] = df["country"].fillna("Unknown")
df["speaker"] = df["speaker"].fillna("Unknown")
df["post"] = df["post"].fillna("Unknown")

print("Shape:", df.shape)
print("Nulls:", df.isnull().sum().sum())
print("\nVoorbeeld:")
print(df.head(3).to_string())

# Opslaan
OUTPUT_PATH = BASE_DIR / "ungdc_clean.parquet"
df.to_parquet(OUTPUT_PATH, index=False)
print(f"\nOpgeslagen naar: {OUTPUT_PATH}")
```

```python
# Maak een schone iso -> landnaam mapping
iso_country_map = (
    df[df["country"] != "Unknown"]
    .groupby("iso")["country"]
    .agg(lambda x: x.value_counts().index[0])  # meest voorkomende naam per iso
    .reset_index()
    .rename(columns={"country": "country_clean"})
)

print(f"Unieke ISO codes: {len(iso_country_map)}")
print(iso_country_map.head(10).to_string())
```

```python
# Voeg schone landnaam toe aan dataframe
df = df.merge(iso_country_map, on="iso", how="left")
df["country_clean"] = df["country_clean"].fillna("Unknown")

print("Unieke schone landen:", df["country_clean"].nunique())
print(df[["iso", "country", "country_clean"]].drop_duplicates("iso").head(10).to_string())
```

```python
# Opslaan
df.to_parquet(BASE_DIR / "ungdc_clean.parquet", index=False)
print("Opgeslagen.")
```
