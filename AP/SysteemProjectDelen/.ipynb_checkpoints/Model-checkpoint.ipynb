```python
import os
import re
import pandas as pd
from pathlib import Path
```

```python
SPEECHES_DIR = Path(r"C:\Users\olive\Documents\SysteemProject\dataverse_files\dataverse_files\UNGDC_1946-2025")
METADATA_PATH = Path(r"C:\Users\olive\Documents\SysteemProject\dataverse_files\dataverse_files\Speakers_by_session.xlsx")

# Load and inspect metadata first
metadata_df = pd.read_excel(METADATA_PATH)
metadata_df = metadata_df.drop(columns=["Unnamed: 6"])
metadata_df.columns = ["year", "session", "iso", "country", "speaker", "post"]
print(metadata_df.dtypes)
metadata_df.head()
```

```python
pattern = re.compile(r"([A-Z]{3})_(\d+)_(\d{4})\.txt")
records = []

for fpath in SPEECHES_DIR.rglob("*.txt"):
    m = pattern.match(fpath.name)
    if not m:
        continue
    iso, session, year = m.group(1), int(m.group(2)), int(m.group(3))
    with open(fpath, "r", encoding="utf-8", errors="replace") as f:
        text = f.read().strip()
    records.append({"iso": iso, "session": session, "year": year, "text": text})

speeches_df = pd.DataFrame(records)
print(speeches_df.shape)
speeches_df.head()
```

```python
df = speeches_df.merge(metadata_df, on=["iso", "session", "year"], how="left")
print(df.shape)
print(df.isnull().sum())
df.head()
```

```python
df = df.drop(columns=["text_len"])
df = df.drop_duplicates(subset=["iso", "session", "year"])
print(df.shape)

df.to_parquet(
    r"C:\Users\olive\Documents\SysteemProject\dataverse_files\dataverse_files\ungdc_merged.parquet",
    index=False
)
print("Saved.")
```

```python
print(df["year"].min(), "–", df["year"].max())
print(f"Unique countries: {df['iso'].nunique()}")
print(f"Unique sessions: {df['session'].nunique()}")
df.groupby("year").size().describe()
```

```python
BOILERPLATE = [
    r"\._[^\s]+",                        # any remaining mac metadata artifacts
    r"mr\.?\s*president",
    r"madam\s*president",
    r"mr\.?\s*secretary[\s\-]*general",
    r"i wish to congratulate",
    r"allow me to",
    r"it is an hono(u)?r",
    r"on behalf of",
    r"i have the hono(u)?r",
    r"i should like to",
    r"my delegation",
]

def clean_speech(text: str) -> str:
    text = text.lower()
    for phrase in BOILERPLATE:
        text = re.sub(phrase, " ", text, flags=re.IGNORECASE)
    text = re.sub(r"\s+", " ", text)      # collapse whitespace
    text = re.sub(r"[^\w\s]", " ", text)  # remove punctuation
    return text.strip()

# Test on one speech before applying to everything
sample = df.loc[df["iso"] == "ARG", "text"].iloc[0]
print("BEFORE:\n", sample[:500])
print("\nAFTER:\n", clean_speech(sample)[:500])
```

```python
df["text_clean"] = df["text"].apply(clean_speech)
print("Done.")
print(df["text_clean"].str.len().describe())
```

```python
df.to_parquet(
    r"C:\Users\olive\Documents\SysteemProject\dataverse_files\dataverse_files\ungdc_merged.parquet",
    index=False
)

print("Saved with clean text.")
```

```python
from bertopic import BERTopic
from sklearn.feature_extraction.text import CountVectorizer

# Vectorizer that strips stopwords — this fixes the keyword problem
vectorizer = CountVectorizer(stop_words="english", min_df=2)

sample_df = df.sample(500, random_state=42)
docs = sample_df["text_clean"].tolist()

model = BERTopic(
    language="english",
    vectorizer_model=vectorizer,
    verbose=True
)
topics, probs = model.fit_transform(docs)

print(f"Topics found: {model.get_topic_info().shape[0] - 1}")
print(f"Outliers: {topics.count(-1)} / {len(topics)}")
model.get_topic_info().head(10)
```

```python
# Full BERTopic run op alle speeches + opslaan

import re
from pathlib import Path

from sklearn.feature_extraction.text import CountVectorizer
from bertopic import BERTopic
from umap import UMAP
from hdbscan import HDBSCAN

# Zorg dat BASE_DIR goed staat
BASE_DIR = Path(r"C:\Users\olive\Documents\SysteemProject\dataverse_files\dataverse_files")

OUTPUT_TOPICS_PATH = BASE_DIR / "ungdc_topics.parquet"
MODEL_PATH = BASE_DIR / "bertopic_model"

# Extended stopwords — generic UN speech terms that appear everywhere
un_stopwords = [
    "nations", "united", "international", "world", "countries", "country",
    "general", "assembly", "delegation", "united nations", "session",
    "president", "secretary", "government", "states", "global",
    "also", "must", "would", "may", "one", "many", "new", "said"
]

extra_stopwords = [
    # Te algemene UN/politieke termen
    "peace", "development", "people", "peoples", "security",
    "economic", "political", "social",
    
    # Let op: human/rights kun je eruit halen, maar dan wordt "human rights" minder zichtbaar.
    # Laat ze erin als je human-rights topics belangrijk vindt.
    # "human", "rights",

    "organization", "organizations", "cooperation", "efforts",
    "support", "need", "important", "continue", "work", "achieve",
    "future", "common", "community", "order", "principles",
    
    # Diplomatieke boilerplate
    "congratulate", "congratulations", "occasion", "honour", "honor",
    "pleasure", "express", "extend", "warmest", "permit", "sir",
    "taking", "floor", "address", "statement", "remarks",
    
    # Algemene werkwoorden/woorden die weinig topicwaarde hebben
    "let", "like", "make", "made", "take", "taken", "give", "given",
    "shall", "can", "cannot", "today", "first", "time", "years",
    
    # UN-specifiek algemeen
    "assembly", "session", "delegation", "general", "secretary",
    "president", "charter", "resolution", "resolutions",
    "member", "members", "state", "states",

    # Extra ruis uit jouw output
    "council", "year", "years", "problem", "problems",
    "body", "election", "elected", "ambassador",
    "minister", "ministers", "heads", "majesties",
    "mr", "mrs", "ms"
]

un_stopwords = list(set(un_stopwords + extra_stopwords))

# Filter UN-specific stopwords from docs
def remove_un_stopwords(text):
    for word in un_stopwords:
        text = re.sub(rf"\b{re.escape(word)}\b", " ", text, flags=re.IGNORECASE)
    return re.sub(r"\s+", " ", text).strip()

# Gebruik alle speeches
full_df = df.copy().reset_index(drop=True)

print("Total speeches:", len(full_df))

docs = full_df["text_clean"].fillna("").tolist()
docs_filtered = [remove_un_stopwords(d) for d in docs]

print("Documents prepared:", len(docs_filtered))

# Vectorizer: gebruikt losse woorden en woordparen zoals "climate change"
vectorizer = CountVectorizer(
    stop_words="english",
    min_df=5,
    ngram_range=(1, 2)
)

# Voor full dataset mag min_cluster_size hoger dan bij 500 sample
umap_model = UMAP(
    n_neighbors=15,
    n_components=5,
    min_dist=0.0,
    metric="cosine",
    random_state=42
)

hdbscan_model = HDBSCAN(
    min_cluster_size=30,
    metric="euclidean",
    cluster_selection_method="eom",
    prediction_data=True
)

model = BERTopic(
    vectorizer_model=vectorizer,
    umap_model=umap_model,
    hdbscan_model=hdbscan_model,
    verbose=True
)

topics, probs = model.fit_transform(docs_filtered)

# Voeg resultaten toe aan dataframe
full_df["topic"] = topics

# Maak simpele automatische topic labels
topic_info = model.get_topic_info()
topic_name_map = dict(zip(topic_info["Topic"], topic_info["Name"]))

full_df["topic_label"] = full_df["topic"].map(topic_name_map)

# Opslaan dataframe met topics
full_df.to_parquet(OUTPUT_TOPICS_PATH, index=False)

# Opslaan BERTopic model
model.save(str(MODEL_PATH), serialization="pickle")

print("Done.")
print(f"Topics found: {topic_info.shape[0] - 1}")
print(f"Outliers: {topics.count(-1)} / {len(topics)}")
print("Saved topic dataframe to:", OUTPUT_TOPICS_PATH)
print("Saved BERTopic model to:", MODEL_PATH)

topic_info.head(20)
```

```python
import pandas as pd
from pathlib import Path

BASE_DIR = Path(r"C:\Users\olive\Documents\SysteemProject\dataverse_files\dataverse_files")

df_topics = pd.read_parquet(BASE_DIR / "ungdc_topics.parquet")

print(df_topics.shape)
print(df_topics.columns)
print(df_topics[["iso", "country", "year", "topic", "topic_label"]].head())
print("Aantal topics:", df_topics["topic"].nunique())
print("Outliers:", (df_topics["topic"] == -1).sum(), "/", len(df_topics))
```

```python
df_topics = df_topics.drop_duplicates(subset=["iso", "session", "year"]).reset_index(drop=True)

print(df_topics.shape)
print("Aantal topics:", df_topics["topic"].nunique())
print("Outliers:", (df_topics["topic"] == -1).sum(), "/", len(df_topics))

df_topics.to_parquet(BASE_DIR / "ungdc_topics.parquet", index=False)
print("Deduplicated and saved.")
```

```python
valid = df_topics[df_topics["topic"] != -1].copy()

# 1. Topic counts
topic_counts = (
    valid.groupby(["topic", "topic_label"])
    .size()
    .reset_index(name="count")
    .sort_values("count", ascending=False)
)

topic_counts.to_csv(BASE_DIR / "topic_counts.csv", index=False)

# 2. Country-topic distribution
country_topic = (
    valid.groupby(["iso", "country", "topic", "topic_label"])
    .size()
    .reset_index(name="count")
)

country_total = (
    valid.groupby(["iso", "country"])
    .size()
    .reset_index(name="total_speeches")
)

country_topic = country_topic.merge(country_total, on=["iso", "country"])
country_topic["share"] = country_topic["count"] / country_topic["total_speeches"]

country_topic.to_csv(BASE_DIR / "country_topic_distribution.csv", index=False)

# 3. Topic-year distribution
topic_year = (
    valid.groupby(["year", "topic", "topic_label"])
    .size()
    .reset_index(name="count")
)

year_total = (
    valid.groupby("year")
    .size()
    .reset_index(name="total_speeches")
)

topic_year = topic_year.merge(year_total, on="year")
topic_year["share"] = topic_year["count"] / topic_year["total_speeches"]

topic_year.to_csv(BASE_DIR / "topic_year_distribution.csv", index=False)

print("Saved analysis files.")
print("topic_counts:", topic_counts.shape)
print("country_topic:", country_topic.shape)
print("topic_year:", topic_year.shape)

topic_counts.head(20)
```
